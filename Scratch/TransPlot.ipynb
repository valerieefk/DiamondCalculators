```python
import DiamondCalcAll as dca 
import numpy as np
import matplotlib.pyplot as plt
```

```python
diamon_thick45 = 45
diamon_thick40 = 40
diamon_thick10 = 20
Pt_metal_thick = 30
Al_metal_thick = 100

flux = 1
energy_graph = 0
pt45curve = 0
pt10curve = 0
al45curve = 0
al10curve = 0
EoffSet = 3000
for val in range (300):
    energy = (val +1) * 100 + EoffSet
    pt45 = dca.MAdevice_perf_pt (diamon_thick45, Pt_metal_thick,energy, flux)
    transDev = 1-(pt45[2]+pt45[3])
    pt45curve = np.append(pt45curve,np.absolute(transDev))
    energy_graph = np.append(energy_graph,energy)

pt10curve = 0

for val in range (300):
    energy = (val +1) * 100 + EoffSet
    pt10 = dca.MAdevice_perf_pt (diamon_thick10, Pt_metal_thick, energy, flux)
    transDev = 1-(pt10[2]+ pt10[3])
    pt10curve = np.append(pt10curve,np.absolute(transDev))

for val in range (300):
    energy = (val +1) * 100 + EoffSet
    al45 = dca.MAdevice_perf_al (diamon_thick45, Al_metal_thick, energy, flux)
    transDev = 1-(al45[2]+ al45[3])
    al45curve = np.append(al45curve,np.absolute(transDev))

for val in range (300):
    energy = (val +1) * 100 + EoffSet
    al10 = dca.MAdevice_perf_al (diamon_thick10, Al_metal_thick, energy, flux)
    transDev = 1-(al10[2] + al10[3])
    al10curve = np.append(al10curve,np.absolute(transDev))

#print(pt45curve)

plt.plot(energy_graph[1:], pt45curve[1:], label = "Pt 45 micron Diamond",color="black")
plt.plot(energy_graph[1:],pt10curve[1:], label = "Pt 20 micron Diamond", color="gray", linestyle = "--")
plt.plot(energy_graph[1:], al45curve[1:], label = "Al 45 micron Diamond")
plt.plot(energy_graph[1:],al10curve[1:], label = "Al 20 micron Diamond", color="green", linestyle = "--")
plt.title("X-ray Transmission Vs. Energy for Beam Monitors")
plt.legend()
plt.savefig("/Users/benjaminmartin/Downloads/DiamondCurve.png")
plt.show()
```
